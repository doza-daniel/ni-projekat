```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
```

```python
file_name = 'COVID_19_PPI/BIOGRID-PROJECT-covid19_coronavirus_project-INTERACTIONS-4.4.210.tab3.txt'
```

```python
df = pd.read_csv(file_name, sep='\t')
df.rename(mapper=lambda col: col.lstrip('#'), axis=1, inplace=True)
df = df.applymap(lambda x: np.nan if x == "-" else x)
df.dropna(subset=['Entrez Gene Interactor A', 'Entrez Gene Interactor B'], inplace=True)
df.dropna(axis=1,inplace=True)
df.head()
```

```python
cols = [
    'Entrez Gene Interactor A',
    'Entrez Gene Interactor B',
    'BioGRID ID Interactor A',
    'BioGRID ID Interactor B',
    'Official Symbol Interactor A',
    'Official Symbol Interactor B'
]

fd = pd.DataFrame(data=df[cols].to_numpy(), columns=cols, index=list(df['BioGRID Interaction ID']))
fd
```

```python
G = nx.DiGraph()

for _, (interaction_id, a_id, b_id, a_symbol, b_symbol) in df.iterrows():
    G.add_node(a_symbol, id=a_id)
    G.add_node(b_symbol, id=b_id)
    G.add_edge(a_symbol, b_symbol, interaction_id=interaction_id)
    
scores = nx.pagerank(G, alpha=0.91)
scores_sorted = sorted(scores.items(), key=lambda x: x[1], reverse=True)

node_id = lambda symbol: G.nodes[symbol]['id']

top = pd.DataFrame(
    [[symbol, node_id(symbol)] for (symbol, _) in scores_sorted[:20]],
    columns=['symbol', 'id'])
top
```

```python
subg = G.subgraph(list(top['symbol']))

pos = nx.circular_layout(subg)
    
plt.figure(figsize=(30, 30))

colors = [scores[node] for node in subg.nodes()]

# Iscrtavamo cvorove
nx.draw_networkx_nodes(subg, pos, cmap=plt.cm.Reds, node_color=colors, node_size=3000, node_shape='h')

# Iscrtavamo oznake cvorova
nx.draw_networkx_labels(subg, pos, font_color='k', font_size=12, font_weight='bold')

# Iscrtavamo grane
nx.draw_networkx_edges(subg, pos, edgelist=subg.edges(), width=1, edge_color='black')

# Iskljucujemo ose
plt.axis('off')

# I prikazujemo graf
plt.show()
```
